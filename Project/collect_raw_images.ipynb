```python
from icrawler.builtin import BingImageCrawler
import os
import unicodedata
```

```python
RAW_DIR = "raw_images"
os.makedirs(RAW_DIR, exist_ok=True)
```

```python
def normalize_query(query):
    return unicodedata.normalize("NFKD", query).encode("ascii", "ignore").decode("ascii")

def scrape_google_images(query, max_images=200):
    # Note: on conserve le nom de fonction pour compatibilité avec le reste du notebook
    query_folder = normalize_query(query.lower().replace(" ", "_"))
    save_dir = os.path.join(RAW_DIR, query_folder)
    os.makedirs(save_dir, exist_ok=True)

    before_count = len([
        f for f in os.listdir(save_dir)
        if os.path.isfile(os.path.join(save_dir, f))
    ])

    crawler = BingImageCrawler(
        storage={'root_dir': save_dir},
        downloader_threads=4
    )

    try:
        crawler.crawl(
            keyword=normalize_query(query),
            max_num=max_images,
            file_idx_offset='auto',
            max_idle_time=8
        )

        after_count = len([
            f for f in os.listdir(save_dir)
            if os.path.isfile(os.path.join(save_dir, f))
        ])
        downloaded_now = max(0, after_count - before_count)
        print(f"  -> +{downloaded_now} nouvelles images (total dossier: {after_count})")
        return downloaded_now
    except Exception as e:
        print(f"  -> échec du crawl: {e}")
        return 0
```
